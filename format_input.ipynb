```python
input = pd.read_csv(r'\\allen\programs\celltypes\workgroups\mousecelltypes\SarahWB\ccf_slice_registration\slices_with_fiducials.csv')
input
```

```python
input = input.rename(columns={'slide_specimen_id' : 'slice_id', 'specimen_name' : 'Specimen Name'})
input
```

```python
specimen_names = input['Specimen Name']
specimen_ids = []
for name in specimen_names:
    try:
        id = get_id_by_name(name.strip())
        specimen_ids = specimen_ids + [id]
    except:
        print('could not get id for ', name)
        specimen_ids = specimen_ids + [None]
```

```python
input['Cell Specimen Id'] = specimen_ids
```

```python
input = input[['slice_id', 'Cell Specimen Id', 'Specimen Name']]
input
```

```python
input = input.dropna()
```
